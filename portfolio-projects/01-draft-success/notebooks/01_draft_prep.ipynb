```python
import pandas as pd

df = pd.read_csv('../data/data_prep/2019-22_draft.csv')
```

```python
print(df.describe())
df
```

```python
df = df.rename(columns={"Drafted (tm/rnd/yr)": "Drafted"})
```

```python
df.columns = [col.lower() for col in df.columns]
df.head()
```

```python
def parse_height(height_str):
    if pd.isna(height_str) or height_str == 'Ht':
        return None
    try:
        feet, inches = map(int, height_str.split("-"))
        return feet + inches / 12.0
    except:
        return None

df["ht_num"] = df["ht"].apply(parse_height)
```

```python
df["round"] = df["drafted"].str.extract("/ (\d+)", expand=False)
```

```python
draft = df.dropna(subset='drafted')
draft
```

```python
draft = draft.drop(columns=['drafted','player-additional','college','ht'], axis=1)
draft
```

```python
draft.to_csv(
    "2019-22_drafted.csv",
    index=False,
    header=True,
    sep=",",
    na_rep="NA",
    quotechar='"',
    encoding="utf-8",
)
```

```python
df_av = pd.read_csv(
    "../data/data_prep/2019-22_av.csv"
)
df_av.head()
```

```python
df_av.columns = [col.lower() for col in df_av.columns]
```

```python
full_df = draft.merge(df_av, on='player', how='left')
full_df.head()
```
